```python
#Importing libraries which are required for the task 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Reading data from the given url
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print ("data imported successfully ")
data.head()
```

```python
#Ploting the distribution of scores
data.plot(x="Hours" , y = "Scores" , style = "o" "r")
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
#Ploting regression line
x = data.Hours
y = data.Scores

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc , x))
plt.scatter(x ,y , c = '#f7230f')
plt.plot(x , mymodel )
plt.show()
std_err
```

```python
#Creating data for calcuating errors
test_x = x[20:]
test_y = y[20:]

predict_y = myfunc(test_x)
df = pd.DataFrame({'Actual' : test_y , 'Predicted' : predict_y})
df
```

```python
#Calculating Errors
from sklearn import metrics
print('Mean Absolute Error : ' , metrics.mean_absolute_error(test_y ,predict_y))
print('Mean Squared Error : '  , metrics.mean_squared_error(test_y ,predict_y))
print('Root Mean squared Error : ' , np.sqrt(metrics.mean_absolute_error(test_y ,predict_y)))
```

```python
#Prediction of Score for 9.25 hours
hours = 9.25
myscore = myfunc(hours)
print("No of Hours : {}".format(hours))
print("Predicted Score : {}".format(myscore))
```
